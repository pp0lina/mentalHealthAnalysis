# K-means clustering

### 1. Load the data from the previous step

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import joblib
import pandas as pd
import numpy as np
import joblib

# Load the data
X_pca = np.load('X_pca.npy')
df = pd.read_csv('aligned_data.csv')

# Check data shape
print(f"Loaded X_pca with shape: {X_pca.shape}")
```

### 2. Determine the optimal number of clusters using the Elbow Method

```python
sse = []
silhouette_scores = []
cluster_range = range(2, 41)

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_pca)
    
    sse.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_pca, kmeans.labels_))
```

### 3. Generate the elbow plot

```python
plt.figure(figsize=(20, 5))
plt.plot(cluster_range, sse, marker='o')
plt.title('Elbow method: optimal number of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.grid()
plt.show()
```

## Plot silhouette scores

```python
plt.figure(figsize=(20, 5))
plt.plot(cluster_range, silhouette_scores, marker='o', color='purple')
plt.title('Silhouette score vs number of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.grid()
plt.show()
```

## Fit K-Means with the optimal number of clusters based on the above analysis

```python
optimal_k = 5
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit_predict(X_pca)
```

## Save the result

```python
cluster_labels = kmeans.labels_
df['Cluster'] = cluster_labels

df.to_csv('data_with_clusters.csv', index=False)

np.save('cluster_labels.npy', cluster_labels)
joblib.dump(kmeans, 'kmeans_model.pkl')
```
