# Visualization

## 1. Load the data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df_with_clusters = pd.read_csv('data_with_clusters.csv')
print(df_with_clusters.columns)
```

## 2. Perform PCA to reduce to 2D and plot the clusters

```python
from sklearn.decomposition import PCA

X = df_with_clusters.drop(columns=['Cluster'])

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

pca_df = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
pca_df['Cluster'] = df_with_clusters['Cluster']

plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=pca_df, palette='viridis', s=100, alpha=0.7)
plt.title('Cluster Visualization in 2D PCA Space')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

pca_3d = PCA(n_components=3).fit_transform(df_with_clusters.drop(columns=['Cluster']))
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(pca_3d[:, 0], pca_3d[:, 1], pca_3d[:, 2], c=df_with_clusters['Cluster'], cmap='viridis', s=50, alpha=0.7)
legend = ax.legend(*scatter.legend_elements(), title="Cluster")
ax.add_artist(legend)
ax.set_title('Cluster separation using 3D PCA')
plt.show()
```

## 3. Perform t-SNE to reduce the data to 2D and plot the clusters

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

tsne_df = pd.DataFrame(X_tsne, columns=['tSNE1', 'tSNE2'])
tsne_df['Cluster'] = df_with_clusters['Cluster']

plt.figure(figsize=(8, 6))
sns.scatterplot(x='tSNE1', y='tSNE2', hue='Cluster', data=tsne_df, palette='viridis', s=100, alpha=0.7)
plt.title('Cluster Visualization in 2D t-SNE Space')
plt.show()
```

## 4. Compute the correlation matrix on the modified dataframe and create the heatmap

```python
# Shorten column names
df_with_clusters.columns = [col.split()[0] for col in df_with_clusters.columns]

corr_matrix = df_with_clusters.drop(columns=['Cluster']).corr()

plt.figure(figsize=(20, 14))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Correlation heatmap of features', fontsize=16)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

## 5. Build pairplots for selected columns

```python
selected_columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5',
                    'Cluster']

df_selected = df_with_clusters[selected_columns]

sns.pairplot(df_selected, hue='Cluster', diag_kind='kde', palette='viridis', corner=True)
plt.suptitle('Pair Plot of Clusters', y=1.02)
plt.show()
```

## 6. Build a box plot for cluster comparison

```python

plt.figure(figsize=(8, 6))
sns.boxplot(x='Cluster', y='PC1', hue='Cluster',
            data=df_with_clusters, palette='viridis',
           legend=False)
plt.title('Impact of Mental Health on Work Across Clusters')
plt.show()
```

## 7. Build a radar chart for each cluster

```python
def radar_chart(data, cluster_labels):
    cluster_means = data.groupby(cluster_labels).mean()
    categories = cluster_means.columns
    num_vars = len(categories)
    
    for cluster_idx in cluster_means.index:
        values = cluster_means.loc[cluster_idx].values.flatten().tolist()
        values += values[:1]
        
        angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
        angles += angles[:1]
        
        plt.figure(figsize=(6, 6))
        ax = plt.subplot(111, polar=True)
        plt.xticks(angles[:-1], categories, color='grey', size=8)
        ax.plot(angles, values, linewidth=1, linestyle='solid', label=f'Cluster {cluster_idx}')
        ax.fill(angles, values, alpha=0.3)
        plt.title(f'Cluster {cluster_idx} Characteristics', size=15, y=1.1)
        plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
        plt.show()

radar_chart(df_with_clusters.drop(columns='Cluster'), df_with_clusters['Cluster'])
```
